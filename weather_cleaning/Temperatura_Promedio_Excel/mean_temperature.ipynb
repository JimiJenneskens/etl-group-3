# Mean or anual average temperature

In this notebook 20 excel files are read as panda's dataframes. For each dataframe, which corresponds to a year, the columns that contain the state and the anual average temperature (°C) are selected and coherced into a new pandas dataframe.  Then, a colum named year is added. Each dataframe is concatenated with the last one, ending up with a dataframe with three columns and 640 observations. This is achieved through a for loop.

Data is provided by CONAGUA, and it is available at: https://datos.gob.mx/busca/dataset/temperatura-promedio-excel

```python
#Call the dependencies
import pandas as pd
```

```python
#Let's create a list of the years of interest
years =  [2000+i for i in range(20)]
print(years)
```

```python
#Create an empty dataframe
avg_temp = pd.DataFrame(columns = ['state', 'average_temperature', 'year', 'id_state']) 
avg_temp
```

```python
#Createa loop to read the excel file for each year, tranform and concatenete to main df
#Generate a loop to read the file and combine it in a unique df
for year in years:
    #Create the path to each excel file
    path =f"{year}Tmed.xlsx"
    #Read the excel file
    prueba = pd.read_excel (path)
    #Drop national observation
    prueba.drop(prueba.tail(1).index, inplace = True) 
    #Select the columns of interest as a list
    data = [prueba["ENTIDAD"], prueba["ANUAL"]]
    #Cretae the name of the columns
    headers = ["state", "average_temperature"]
    #Coherce in a df
    aux = pd.concat(data, axis=1, keys=headers)
    #Add a column with the year 
    aux["year"] = [year] * 32
    aux.sort_values(by=['state'], inplace=True)
    #Lets create a numeric id
    seq_s = [i+1 for i in  range(32)]
    #Lets add it to the df
    aux['id_state'] = seq_s
    #Concatenate in max_tem
    avg_temp = pd.concat([avg_temp, aux])
```
