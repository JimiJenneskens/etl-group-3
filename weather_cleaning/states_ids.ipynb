# Generating an id for the states.

In this notebook it is created a dataframe that is going to contain the abreviation of the states name, the name and a number that is going to work as a primary key.
This is necesarry, cince in all of our data the names of the states are writen differet.


```python
import pandas as pd
```

```python
#Lets call the rain df
states = pd.read_csv("states_abbreviation.csv")
states.head()

```

```python
#lETS order by name
states.sort_values(by=['state'], inplace=True)
states
```

```python
#Lets create a numeric id
seq_s = [i+1 for i in  range(32)]
#Lets add it to the df
states['id_state'] = seq_s
states = states[['id_state', 'state', 'abbr_state']]
states.rename(columns={"id_state":"id"}, inplace=True)
#call the df
states
```

```python
#Lets save it
states.to_csv('../data/states_ids.csv', index = False)
```
