# Merging weather variables.

In this notebook four csv are joined as one single table containing four weather avriable at a state level from 2000 to 2019, which are:

Precipitation in mm annualy

Maximum temperature in °C annualy

Minimum temperature in °C annualy

Mean temperature in °C annualy

```python
#Let's import the dependencies
import pandas as pd
```

```python
#Create the paths

prain = "precipitation.csv"
pmaxt = "maximum_temp.csv"
pmint = "minimum_temp.csv"
pavgt = "average_temp.csv"
```

```python
#Read the files
rain_df = pd.read_csv(prain, delimiter=",", encoding='utf8')
rain_df.head()
```

```python
temp_max_df = pd.read_csv(pmaxt,delimiter=",", encoding='utf8')
temp_max_df.head()
```

```python
temp_min_df = pd.read_csv(pmint,delimiter=",", encoding='utf8')
temp_min_df.head()
```

```python
temp_mean_df = pd.read_csv(pavgt, delimiter=",", encoding='utf-8')
temp_mean_df.head()
```

```python
#Lets check the df's
rain_df.describe()
```

```python
temp_max_df.describe()
```

```python
temp_min_df.describe()
```

```python
temp_mean_df.describe()
```

```python
#Lets merge in a big df
#Merge rain and mean temperature
aux = pd.merge(rain_df, temp_mean_df, how="outer", on=["state", "year","id_state"])
print(aux.info)
aux.head()
```

```python
#Merge aux with min temp
aux2 = pd.merge(aux, temp_min_df, how = "outer", on=["state", "year","id_state"])
print(aux2.info)
aux2.head()
```

```python
#Merge aux 2 with maximum temperature
weather_df = pd.merge(aux2, temp_max_df, how = "outer", on=["state", "year","id_state"])
print(weather_df.info)
weather_df.head()
```

```python
weather_df = weather_df[["id_state","state", "year", "precipitation", "average_temperature", "minimum_temperature", "maximum_temperature"]]
```

```python
weather_df.rename(columns={"state":"state_w"}, inplace=True)
```

```python
weather_df.head()
```

```python
#Lets save it
weather_df.to_csv('../data/weather.csv', index = False, encoding="utf-8")
```
