# Average rainfall

In this notebook precipitation data is ordered. The data for year 2000 to 2019 is embeded in 20 excel files, one for each year. 
Data is available at CONAGUA web page: https://datos.gob.mx/busca/dataset/precipitacion .
Each excel file contains 14 columns, the first one keeps track of the stae and the next 12 ones has the precipitation data for each month in $\frac{mm}{m^{3}}$. The last colum contains the anual average rainfall. 
The excel files has 33 rows, one row for each Mexico's state and a the national data.

```python
#Call the dependencies

import pandas as pd
  
```

```python
#Generate a list of years
years =  [2000+i for i in range(20)]
print(years)
```

```python
#Create an empty df
rain = pd.DataFrame(columns = ['state', 'precipitation', 'year','id_state']) 
rain
```

```python
#Generate a loop to read the file and combine it in a unique df
for year in years:
    path =f"{year}Precip.xls"
    prueba = pd.read_excel (path)
    prueba.drop(prueba.tail(1).index, inplace = True) 
    data = [prueba["ENTIDAD"], prueba["ANUAL"]]
    headers = ["state", "precipitation"]
    aux = pd.concat(data, axis=1, keys=headers)
    aux["year"] = [year] * 32
    aux.sort_values(by=['state'], inplace=True)
    #Lets create a numeric id
    seq_s = [i+1 for i in  range(32)]
    #Lets add it to the df
    aux['id_state'] = seq_s
    #concatenateall the df
    rain = pd.concat([rain, aux])
```
