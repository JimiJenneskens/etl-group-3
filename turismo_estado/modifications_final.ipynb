```python
import pandas as pd

# Merge excel files
path = "turismoestado.csv"

dataset_df = pd.read_csv(path)

dataset_df
```

```python
#set ano as index and swap axes
dataset_df = dataset_df.set_index('ano')
dataset_df = dataset_df.swapaxes("index", "columns")

dataset_df.head()
```

```python
#reset index and remove index name
dataset_df.reset_index(level=0, inplace=True)
dataset_df.columns.name = None

dataset_df.head()
```

```python
#reset index again to get id and rename index to state and level_0 to id
dataset_df.reset_index(level=0, inplace=True)
dataset_df.rename(columns={'index': 'state', 'level_0': 'id'}, inplace=True)
dataset_df.head()
```

```python
new_df = dataset_df.iloc[1:]
```

```python
new_df.head()
```

```python
new_df.sort_values(by=['state'], inplace=True)
new_df
```

```python
#Add an id state
seq_s =[i+1 for i in range(32)]
new_df["id_state"]=seq_s
new_df
```

```python
#Let's turn to upper case all the names of the states
new_df['state'] = new_df['state'].str.upper() 
new_df
```

```python
#Lets change the order of the df
#Create a list of year
years = [2003+i for i in range(16)]
print(years)
```

```python
#Lets create an empty df
gdp_df = pd.DataFrame(columns = ['id_state','states', 'tourism_gdp', 'year']) 
gdp_df
```

```python
#Loop to rewrite the df
for year in years:
    syear = f"{year}"
    #Select the columns of interest as a list
    data = [new_df["id_state"], new_df["state"], new_df[year]]
    #Cretae the name of the columns
    headers = ['id_state','states', 'tourism_gdp',]
    #Coherce in a df
    aux = pd.concat(data, axis=1, keys=headers)
    #Add a column with the year 
    aux["year"] = [year] * 32
    #Concatenate in the empty df
    gdp_df = pd.concat([gdp_df, aux])
```
