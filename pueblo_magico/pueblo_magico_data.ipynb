```python
import os
import pandas as pd
import numpy as np
from datetime import timedelta

# Merge excel files
path = "pueblos_magicos.csv"

dataset_df = pd.read_csv(path, encoding = 'unicode_escape')

dataset_df
```

```python
columns_titles = ["municipio","estado", "ano_ingreso"]
dataset_df = dataset_df.reindex(columns=columns_titles)

dataset_df.reset_index(level=0, inplace=True)
dataset_df.rename(columns={'index': 'id', 'municipio': 'town', 'estado': 'state', 'ano_ingreso': 'year_entered'}, inplace=True)
dataset_df
```
