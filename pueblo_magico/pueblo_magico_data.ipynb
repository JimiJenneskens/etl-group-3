```python
import os
import pandas as pd
import numpy as np
from datetime import timedelta

# Merge excel files
path = "pueblos_magicos.csv"

dataset_df = pd.read_csv(path, encoding = 'unicode_escape')

dataset_df
```

```python
columns_titles = ["municipio","estado", "ano_ingreso","id_edo"]
dataset_df = dataset_df.reindex(columns=columns_titles)

dataset_df.reset_index(level=0, inplace=True)
dataset_df.rename(columns={'index': 'id', 'municipio': 'town', 'estado': 'state_mt', 'ano_ingreso': 'year', "id_edo":"id_state"}, inplace=True)
dataset_df
```

```python
#Lest change to upper case the state
  
dataset_df['state_mt'] = dataset_df['state_mt'].str.upper() 
dataset_df
```

```python
dataset_df.to_csv('../data/magic_town.csv', index = False)
```
